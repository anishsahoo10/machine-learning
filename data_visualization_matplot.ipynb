```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
y = x**3
```

```python
#basics
plt.plot(x,y)
#added lables
plt.title("hellow strong coffee")
plt.xlabel("x axis")
plt.ylabel("y axis")
```

# subplots

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,4)
plt.plot(x,y)
plt.subplot(2,2,3)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(x**2,x**2)
```

# object oriented way

```python
#obj
fig = plt.figure(figsize=(10,10),dpi=100) #dept per inch
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.5])
axis2.plot(x,y)
```

# types of plots

```python
plt.scatter(x,y)
```

```python
#histogram
from random import sample
data = sample(range(1,100),10)
plt.hist(data)
```

```python
plt.boxplot(x)
```

# saving plots

```python
fig = plt.figure()
ax =fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("firstplot.png")
```
